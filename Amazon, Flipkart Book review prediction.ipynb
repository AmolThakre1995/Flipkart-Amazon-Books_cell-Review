```python
data=data.drop(['amazon_isbn-10','flipkart_isbn10','flipkart_title','amazon_author'],axis=1)
```

```python
data1=data.rename(columns={'amazon_title':'Book Title','flipkart_author':'Book Author','flipkart_stars':'Flipkart_rating',
            'amazon_reviews count':'amazon_rating counts'})
```

```python
plt.figure(figsize=(16,8))
plt.hist(x=data1['amazon_price'],bins=180,label='Amazon Price',color='r')
plt.hist(x=data1['flipkart_price'],bins=180,label='Flipkart Price',color='c',alpha=0.6)
plt.legend()
plt.xlabel('Price of Book')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart Price')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.hist(x=data1['amazon_rating'],color='k',bins=100,label='Amazon Ratings')
plt.hist(x=data1['Flipkart_rating'],color='r',bins=100,label='Flipkart Rating')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart Rating')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.hist(data1['amazon_rating counts'],edgecolor='r',bins=150,fc=(0,0,1,0.7),label='Amazon Rating',alpha=0.3,range=[0,1000])
plt.hist(data1['flipkart_ratings count'],edgecolor='k',bins=150,fc=(0,0,1,0.7),label='Flipkart Rating',range=[0,1000])
plt.legend()
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart Rating Coounts')
plt.show()
```
